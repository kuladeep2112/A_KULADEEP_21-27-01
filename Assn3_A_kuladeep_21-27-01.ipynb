## Retrieve all NANs in the form of boolean list

```python
x=np.array(df)
```

```python

nans=[]
row,column=x.shape[0],x.shape[1]
print(row,column)
for i in range(row):
    List=[]
    for j in range(column):
        try:
            float(x[i][j])
            if float(x[i][j])>0 or float(x[i][j])<=0:
                List.append(True)
            else:
                List.append(False)
        except ValueError:
            if isinstance(x[i][j],str):
                if x[i][j].isalpha():
                    if x[i][j]=='HURLEY' or x[i][j]=='na':
                        List.append(False)
                    else:
                        List.append(True)
                else:
                    List.append(False)
     
    nans.append(List)
nans
```

## Delete all rows with NaN values

```python
df1=df.copy()
index1=[]
for i in range(len(df)):
    try:
        int(df1['NUM_BEDROOMS'][i])
    except ValueError:
        index1.append(i)
        continue
df1.drop(df1.index[index1], inplace=True)
df1
```

## Replace NaN with previous entry

```python
df_new=df.copy()
for i in range(len(df)):
    try:
        a=float(df_new['NUM_BATH'][i])
        if a>0 or a<=0:
            continue
        else:
            df_new['NUM_BATH'][i]=df_new['NUM_BATH'][i-1]
    except ValueError:
        df_new['NUM_BATH'][i]=df_new['NUM_BATH'][i-1]
df_new
```

## Replace empty values in SQ_FT with mean

```python
index1=[]
sum=0
count=0
for i in range(len(df)):
    try:
        a=int(df_new['SQ_FT'][i])
        sum+=a
        count+=1
    except ValueError:
        index1.append(i)
mean=sum/count
for i in index1:
    df_new["SQ_FT"][i]=mean  
df_new
```

## Sum of all values in column SQ_FT

```python
sum=0
for i in range(len(df)):
    sum+=int(df_new['SQ_FT'][i])
sum   
```
